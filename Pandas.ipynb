```python
# 열단위 추출

# 1개의 열만 추출하면 시리즈를 얻게 되고, 2개 이상의 열을 추출하게 되면
# 데이터 프레임을 얻게 된다.
country_df = df['country']

type(country_df)
```

```python
subset = df[['country','continent','year']]
subset
```

```python
#  행단위 추출
#  loc 
print(df.loc[0])
# 괄호 안에 인덱스값을 넣어주면 행 전체의 데이터를 추출할 수 있다.
print()
print(df.loc[len(df)-1])
print()
# iloc
```

```python
# iloc로 행 추출하기
# loc는 데이터 프레임의 인덱스 번호를 사용하여 추출
# iloc는 데이터 순서를 의미하는 행 번호를 사용한다.

df.iloc[-1]
```

```python
df.iloc[[0,99,999]]
```

```python
# 슬라이싱

df.loc[:,['year','pop']]
```
